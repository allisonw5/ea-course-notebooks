```python
months = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December']
month_abbrs = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 
    'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'
]

for month, month_abbr, month_num in zip(months, month_abbrs, range(1, 13)):
    print(month, month_abbr, month_num)
```

```python
for i in range(3, 8):
    print(i)
```

```python
month_num = 1
for month, month_abbr in zip(months, month_abbrs):
    print(month, month_abbr, month_num)
    month_num += 1
```

```python
counts = list(range(1, 11))
counts.append(14)
print(counts)
```

```python
new_counts = []
for count in counts:
    new_counts.append(count * 2)
    print(new_counts)

print(new_counts)
```

```python
count_sum = 0
for count in counts:
    count_sum += count
    print(count_sum)
count_sum
```

```python
month_info = {}
for month_idx, (month_name, month_abbr) in enumerate(zip(months, month_abbrs)):
    month_num = month_idx + 1
    month_info[month_num] = {'name': month_name, 'abbr': month_abbr}
    print(month_info)
month_info[3]['abbr']
```

```python
for month_num, month_names in month_info.items():
    print(month_num, month_names)
```
